# Oscillation Word Problem

Free Fall Adventures Inc., a company specializing in providing outdoor excitement to customers, wants to provide its clients with the chance to jump off the Quechee Gorge bridge 51 meters over the Ottaquechee River in Vermont. They hire you to get them the right cord and cut it to the right length. You assume that the mass of people doing the jump will be 80 kg. You want the maximum acceleration of a human being to be 20 m/s2 and you want the closest they get to the surface of the river to be 4 meters. Using both energy and force methods, determine:

(a) the length of the cord you should get for this jump;

(b) the spring constant of the cord;

(c) the maximum speed of the jumper.

### The relationship between $k$, $x$, and energy

There are three types of energy in this system: gravitational potential (GPE), spring potential or elastic potential (EPE), and kinetic (KE).

$GPE = mgh$

$EPE = \frac{1}{2}kx^2$

$KE = \frac{1}{2}mv^2$

Total mechanical energy is the sum of all three. There are no non-conservative forces acting on this system, so at every point, total mechanical energy must be the same. 


Consider the bungeer jumper right before they jump off. The only energy in the system is GPE, which is equal to $mgh$. Later, when the person jumps, no work is done, so the total energy is constant and equal to $mgh$. 

Now consider the bungee jumper when they are at the very bottom of the jump. They have an instantaneous velocity of 0, so $KE = 0$. $GPE$ is $mgh$, and $EPE$ is the remainder of $TME$.

The jumper is supposed to stop at a height of 4 meters above the river. If the river is set to be the "ground" ($h = 0$), then the energy at either point can be solved

$TME = GPE_{top}$

$TME = GPE{bottom} + EPE{bottom}$

These two equations are equivalent, and cna be used to get an expression for EPE.

$GPE_{top} = GPE{bottom} + EPE{bottom}$

$mgh_{top} = mgh_{bottom} + \frac{1}{2}k(x_{max})^2$

$80(9.8)(51) = 80(9.8)(4) + \frac{1}{2}k(x_{max})^2$

$36848 = \frac{1}{2}k(x_{max})^2$

This expression gives the EPE at the bottom and the relationship between $k$ and the maximum displacement of the cord.

$73696 = k(x_{max})^2$

### The relationship between $k$, $x$, and acceleration

The net force on the spring at any time is $F_{spring} - F_g$, or $-kx - mg$. 

Following Newton's laws, $a = \frac{F_{net}}{m}$, or in this case, $a = \frac{-kx - mg}{m}$

The maximum desired acceleration is $20\ m/s$.

Sof $a\le20$, or $\frac{-kx - mg}{m}\le20$

Rearranging this equation

$\frac{-kx}{m}-g\le20$

$\frac{-kx}{80}-9.8\le20$

$-kx\le2384$

The earlier equation for $EPE$ solved for $x_{max}$, the furthest possible displacement, in terms of energy. This can now be solves in terms of acceleration too. As $x$ gets larger in magnitude, the maximum acceleration increases, so given the acceleration limit of $20 m/s^2$,

$-kx_{max}=2384$

This again gives the relationship between $k$ and max displacement of the cord, this time in the context of acceleration instead of energy.

### Solving the system of equations

We now have a system of equations relating $k$, $x$, and some numbers.

$k(x_{max})^2 = 73696$

$-kx_{max}=2384$

After manipulating the second equation to look more like the first one (by multiplying each side by $-x_{max}$)

$k(x_{max})^2=-2384x_{max}$

We can set them equal to each other.

$k(x_{max})^2 = k(x_{max})^2$

$73696 = -2384x_{max}$

$-30.91275 = x_{max}$

This means that the maximum displacement, or the farthest the cord can stretch, is $-30.9\ m$ (or $30.9\ m$ downwards).

For the person to go from the bridge at $51\ m$ high to $4\ m$ above the river, the total stretched length of the cord needs to be 47 m. The stretched length is euqal to the unstretched length plus $30.9\ m$.

$l + |x_{max}| = 47$

$l = 16.1$

#### The unstretched length of the cord is $16.1\ m$

Now that $x$ has been solved for, one of the 2 earlier equations with a $k$ and $x$, can be used to solve for $k$.

1st equation:

$k(x_{max})^2 = 73696$

$k = \frac{73696}{(x_{max})^2}$

$k = \frac{73696}{(-30.9)^2}$

$k = 77.183$

2nd equation:

$-kx_{max}=2384$

$k= \frac{2384}{-x_{max}}$

$k= \frac{2384}{-(-30.9)}$

$k=77.152$

These answers are almost identical, and the slight disagreement is due to rounding approximations in the displacement distance and acceleration due to gravitation constant. 

#### The spring constant, $k$, of the cord is $77\ N/m$

## Finding velocity

The easiest way to find velocity in this non-constantly accelerating system is via energy. At any point between the top and bottom of the jumper's journey, there is some combination of GPE, EPE, and KE, and the sum of these is the total energy, which stays constant. 

$KE = TME - (GPE + EPE)$

Thus, KE is maximized when the sum of GPE and EPE are minimized. We know the formula for GPE and EPE, so if we let each of them be a function, we can then construct a third function which equals the sum of them, and solve the optimization problem.

$GPE(h) = mgh$

$EPE(x) = \frac{1}{2}kx^2$ 

The EPE function is in terms of $x$, but $x$ is height-dependent, so it can be rewritten in terms of $h$. If the length of the unstretched cord is $16.1\ m$, and it hangs from a height of $51\ m$, then the bottom of it is $34.9\ m$ above ground, and this is the spring position at rest. $x$ is the difference between rest and current position, so  $x = h - 34.9$

$EPE(h) = \frac{1}{2}k(h - 34.9)^2$ 

$SGE(h) = mgh + \frac{1}{2}k(h - 34.9)^2$

Taking the derivative and finding critical points:

$SGE'(h) = mg + k(h - 34.9)$

$0 = mg + k(h - 34.9)$

$mg = -k(h - 34.9)$

As established earlier, $x = h - 34.9$, so $mg = -kx$

If you remember from our acceleration equation, $F_{net} = -kx - mg$, so if $mg = -kx$, $F_{net} = 0$ and acceleration equals 0. Similar to how a particle's velocity is 0 when it reaches max height (assuming it's thrown up and allowed to fall down), the particles acceleration is 0 when it reaches max speed (assuming it turns around and goes the other direction).

Solving for $x$ at the point of no acceleration:

$80(9.8) = -77x$

$x = -10.18$

So the minimum (GPE + EPE) and maximum KE occurs when the cord is stretched $10.18\ m$ below rest.

Going back to the equation for $KE$,

$KE = TME - (GPE + EPE)$

$KE = 39984 - (80(9.8)(24.72) + \frac{1}{2}77(10.18^2))$

$KE = 16613.67$

Since $KE = \frac{1}{2}mv^2$,

$16613.67 = \frac{1}{2}mv^2$

$\frac{16613.67(2)}{80} = v^2$

$\pm\sqrt{415.34} = v$

$v = \pm20.38$

Since the bungee jumper is falling down, the velocity will be negative.

#### The maximum velocity is $-20.38\ m/s$
